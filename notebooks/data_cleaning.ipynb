# Handling missing data

```python
# realease_date
round(movies.release_date.isna().sum() * 100 / len(movies),2)
```

```python
# Null values represent 16.22%. Since I will need this data in the future, I will drop these rows to ensure a reliable dataset.
movies.dropna(subset = ['release_date'], inplace = True)
movies.info()
```

```python
# There are only 4 missing values on the title column so i will remove them
movies.dropna(subset = ['title'], inplace = True)
```

```python
round(movies.genres.isna().sum() * 100 / len(movies), 2)
```

```python
round(movies.overview.isna().sum() * 100 / len(movies) , 2)
```

```python
# Since missing values in 'genres' and 'overview' represent a large portion of the dataset, I will fill them with a generic value like 'No overview' and 'No genre'.
# movies.overview.fillna('No overview', inplace = True)
# movies.genres.fillna('No genre', inplace = True)
movies.fillna({ 'overview' : 'No overview', 'genres': 'No genre'}, inplace = True)
movies.info()
```

```python
# Converting release_date to datetime and adding a new column 'release_year'
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
movies['release_year'] = movies['release_date'].dt.year
movies.info()
```

```python
movies.describe()
```

```python
# It seems that vote_average and vote_count have values of 0, so I will filter the dataset.
movies = movies[(movies['vote_count'] > 10) & (movies['vote_count'] > 10)]
movies.describe()
```

```python
# The max and min values in release_date and release_year seem to be errors.
movies = movies[(movies['release_year'] >= 1900) & (movies['release_year'] <= 2024)] #This will filter release_date too
movies.describe()
```

```python
# Runtimes of 0 minutes are likely errors, so I will filter them out.
movies = movies[(movies['runtime'] > 0) & (movies['runtime'] <= 300)]
movies.describe()
```

```python
genres_rd = movies['genres']
genres_un = genres_rd.unique()
unique_genres = set()
for genres in genres_un:
    # Split by ',' and strip any extra whitespace
    unique_genres.update([genre.strip() for genre in genres.split(',')])

# Convert the set back to a sorted list (optional)
unique_genres = sorted(unique_genres)

# Create a dictionary with genre_id as keys
genres_dict = {genre_id: genre for genre_id, genre in enumerate(unique_genres, start=1)}

genres_reverse_dict = {v: k for k, v in genres_dict.items()}

# Function to map genres to a dictionary of {genre_id: genre}
def map_genres_to_dict(genre_string):
    genres_list = [genre.strip() for genre in genre_string.split(',')]
    return {genres_reverse_dict[genre]: genre for genre in genres_list if genre in genres_reverse_dict}

genres_rd = genres_rd.apply(map_genres_to_dict)
movies['genres'] = movies['genres'].apply(map_genres_to_dict)
movies['genres'].head()
```

```python
def extract_genre_ids(genre_dict):
    return list(genre_dict.keys())

# Create the movies_genres DataFrame
movies_genres = movies[['id', 'genres']].copy()
movies_genres['genres'] = movies_genres['genres'].apply(extract_genre_ids)

movies_genres.head()
```

```python
movies.drop(columns = ['genres', 'release_date'], inplace = True)
movies.head()
```

```python
genres = pd.DataFrame.from_dict(genres_dict, orient = 'index')
genres
```

```python
genres.rename(columns={ 0: 'genres'}, inplace = True)
genres
```

```python
import sqlalchemy
from dotenv import load_dotenv
import os

load_dotenv()

db_host = os.getenv('DB_HOST')
db_port = os.getenv('DB_PORT')
db_name = os.getenv('DB_NAME')
db_user = os.getenv('DB_USER')
db_password = os.getenv('DB_PASSWORD')

db_url = f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"

engine = sqlalchemy.create_engine(db_url)
connection = engine.connect()
```

```python
movies.to_sql('movies', engine, index=False, if_exists='replace')
genres.to_sql('genres', engine, index= True, index_label = 'id', if_exists='replace')
movies_genres.to_sql('movie_genres', engine, index=False, if_exists='replace')
```

```python
query = """
SELECT 
    m.id AS movie_id,
    m.title,
    m.popularity,
    m.vote_average,
    m.runtime,
    m.release_year,
    array_agg(g.genres) AS genres
FROM 
    movies m
JOIN 
    movie_genres mg
ON 
    m.id = mg.id
JOIN 
    genres g
ON 
    g.id = ANY(mg.genres::int[])
GROUP BY 
    m.id, m.title, m.popularity, m.vote_average, m.runtime, m.release_year
LIMIT 10;
"""
df = pd.read_sql(query, engine)
df
```
